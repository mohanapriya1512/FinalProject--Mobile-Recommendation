```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver import ActionChains
from urllib.parse import urlparse, parse_qs
import pandas as pd
import time
```

```python
# Initialize the Chrome driver

driver = webdriver.Chrome()
driver.maximize_window()
time.sleep(5)

# Product page URL and product ID extraction
url = "https://www.flipkart.com/samsung-galaxy-f55-5g-raisin-black-128-gb/product-reviews/itmf0394cdb68411?pid=MOBHY9N796YGFJY2&lid=LSTMOBHY9N796YGFJY2BSFXRE&marketplace=FLIPKART"
product_id = url.split('pid=')[1].split('&')[0]  # Extract product ID from the URL

# Load the page
driver.get(url)
wait = WebDriverWait(driver,10)
time.sleep(3)

# Define a function to scroll down the page
def scroll_down():
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
   

# Now scrape the data from the current page:
def scrape_data(): 
    reviews_list=[]
    try:
        time.sleep(3)  
        scroll_down()

        review_container = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.col.EPCmJX.Ma1fCG")))
        time.sleep(3)
        for review in review_container:
            mobile_ratings = review.find_element(By.CSS_SELECTOR, "div.XQDdHH.Ga3i8K").text
            mobile_reviews = review.find_element(By.CSS_SELECTOR, "div.ZmyHeo").text
            reviews_list.append( {
                "Product_ID": product_id,
                "Ratings": mobile_ratings,
                "Reviews": mobile_reviews
            })
                 
        #scroll_down()
        time.sleep(3)
    except Exception as e:
        print(f"Error scraping reviews: {e}")
    return reviews_list


    
# Function to navigate to the specific page number
def next_page(page_number):
    try:
        # Dynamically find the page link using the page number's text
        page_num = f'//a[text()="{page_number}"]'
        page_button = driver.find_element(By.XPATH, page_num)
        page_button.click()
        time.sleep(3)  # Wait for the next page to load
    except Exception as e:
        print(f"Error navigating to page {page_number}: {e}")
        return False
    return True


# Scraping multiple pages (from page 1 to 20)
reviews_all_list=[]
for page in range(1, 20):
    print(f"Scraping page {page}")
    reviews_all_list.extend(scrape_data())
    
    if page < 20:  # Only navigate to the next page if not on the last one
        if not next_page(page + 1):
            break

# Convert the collected data into a DataFrame and print it
pd.set_option('display.max_colwidth', None)
df1_samsung_review = pd.DataFrame(reviews_all_list)
df1_samsung_review['Mobile Name']='SAMSUNG Galaxy F55 5G'
#print(df_samsung_review)
df1_samsung_review.to_csv('Samsung_Mobile_reviews.csv',index=False)
print("Scraping complete, data saved to 'Samsung_Mobile_reviews.csv'")

# Close the browser
driver.quit()
```
